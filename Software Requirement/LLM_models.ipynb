```python
import pandas as pd
import ollama

# Read the CSV file
csv_file_path = '/content/Requirement_data_part1.csv'  # Replace with your actual file path
data = pd.read_csv(csv_file_path)

# Assuming the requirements are in a column named 'Requirement'
requirements_column = 'Requirement'  # Update with the actual column name if it's different

# Create a list to store the updated types
updated_types = []

# Iterate over each row in the dataset
for idx, row in data.iterrows():
    # Use the updated prompt
    prompt = (f'Identify the type of requirements for the following statement: "{row[requirements_column]}". '
              'Provide only the best option from the following categories: Functional (F), Availability (A), '
              'Fault Tolerance (FT), Legal (L), Look & Feel (LF), Maintainability (MN), Operational (O), '
              'Performance (PE), Portability (PO), Scalability (SC), Security (SE), Usability (US), Non-Functional Requirement (NFR).')

    # Call the ollama chat API
    response = ollama.chat(model="llama3.2", messages=[{"role": "user", "content": prompt}])

    # Get the response content
    requirement_type = response['message']['content']

    # Append the response to the updated_types list
    updated_types.append(requirement_type)

    # Print output in the desired format
    print(f"Prompt {idx + 1}: {prompt}")
    print(f"Response {idx + 1}: {requirement_type}\n")

# Add the updated types to the DataFrame
data['updated_type'] = updated_types

# Save the updated DataFrame to a new CSV file
output_csv_file_path = '/content/updated_software_requirements.csv'  # Path for the new CSV file
data.to_csv(output_csv_file_path, index=False)

print(f"Updated CSV saved to {output_csv_file_path}")
```
